```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/s?k=playstation+5&crid=8YD08QGC979P&sprefix=playstation+5+%2Caps%2C269&ref=nb_sb_noss_2"
```

```python
# Headers for request
HEADERS = ({'', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent 
```

```python
webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
#Soup Object Containing all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
#Fetch links as list of tag objects
links = soup.find_all("a", attrs={'class':'a-link-normal s-line-clamp-2 s-line-clamp-3-for-col-12 s-link-style a-text-normal'})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
product_list = "https://amazon.in"+link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
# Soup Object containiang all data
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
new_soup.find("span", attrs={"id":'productTitle'}).text.strip()
```

```python
price = new_soup.select_one("span.a-price span.a-offscreen").text
print(price)
```
